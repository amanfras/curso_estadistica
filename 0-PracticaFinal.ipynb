Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.
Así mismo también debería ser NA todos los que no compiencen por 28, ya que estamos con códigos postales de Madrid

El código postal 28002, 28004 y 28051 tienen entradas repetidas. 
Por ejemplo las entradas 28002\n20882 deberían ir dnetro de 28002

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 juncto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

```r
airbnb$CodigoPostal[airbnb$CodigoPostal == ""] <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == "-"] <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '28'] <- NA

airbnb$CodigoPostal[substr(airbnb$CodigoPostal, start = 1, stop = 6)=='Madrid'] <- '28004'
airbnb$CodigoPostal[substr(airbnb$CodigoPostal, start = 1, stop = 2) != '28'] <- NA

airbnb$CodigoPostal[airbnb$CodigoPostal == '28002\n28002'] <- '28002'
airbnb$CodigoPostal[airbnb$CodigoPostal == '28051\n28051'] <- '28051'

airbnb$CodigoPostal[airbnb$CodigoPostal == '280013'] <- '28013'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2815'] <- '28015'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2805'] <- '28005'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2804'] <- '28004'
```

Cual es el precio medio de alquiler de cada uno, la diferencia que hay ¿es estadísticamente significativa?
¿Con que test lo comprobarías?

```r
#media con outlier
aggregate(Precio ~ TipoAlquiler, airbnb, mean)

#comprobamos que siguen una normal 
shapiro.test(airbnb$Precio[airbnb$TipoAlquiler=="Entire home/apt"][1:5000])$p.value
shapiro.test(airbnb$Precio[airbnb$TipoAlquiler=="Private room"][1:5000])$p.value
shapiro.test(airbnb$Precio[airbnb$TipoAlquiler=="Shared room"][1:5000])$p.value

paste("Como los p-valores son muy pequeños rechazamos la hipótesis nula y por lo tanto no se distribuyen como una normal.")
paste("Vamos a comprobar si la diferencia de las medias es estadísticamente significativa mediante el test de Kruskal-Wallis.")

kruskal.test( Precio ~ TipoAlquiler, data=airbnb)

paste("Como el p-valor es muy pequeño rechazamos la hipótesis nula y por lo tanto si hay diferencias estadísticamente significativas entre los precios medios.")
```

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

```r
#comprobamos que siguen una normal 
shapiro.test(airbnb_entire$MetrosCuadrados[airbnb$Barrio=="Goya"])$p.value
shapiro.test(airbnb_entire$MetrosCuadrados[airbnb$Barrio=="Justicia"])$p.value
shapiro.test(airbnb_entire$MetrosCuadrados[airbnb$Barrio=="Recoletos"])$p.value
shapiro.test(airbnb_entire$MetrosCuadrados[airbnb$Barrio=="Sol"])$p.value
shapiro.test(airbnb_entire$MetrosCuadrados[airbnb$Barrio=="Trafalgar"])$p.value

paste("Como tres de los conjuntos no siguen una distribución normal al tener un p-valor menor de 0.05 vamos a realizar el test de Kruskal-Wallis.")

airbnbbarrios <- airbnb_entire[airbnb_entire$Barrio=="Goya" |airbnb_entire$Barrio=="Justicia" |airbnb_entire$Barrio=="Recoletos" |airbnb_entire$Barrio=="Sol" |airbnb_entire$Barrio=="Trafalgar",]
kruskal.test(MetrosCuadrados ~ Barrio, data=airbnbbarrios)

paste("Como el p-valor es mayor que 0.05 aceptamos la hipótesis nula y por lo tanto no hay diferencias estadísticamente significativas entre los metros cuadrados medios.")
```

Se observa que la correlación entre el número de dormitorios y los metros cuadrados es sorprendentemente baja.
¿Son de fiar esos números?

Mediante un histograma o curvas de densidad podemos descartar números que notienen sentido en el dataframe barrio_sol,
para tener una matriz de correlación que tenga mayor sentido.


```r
#Estos números no son de fiar debido a que pueden haber errores en las entradas, al observar todo en hisitograma y curva de densidad
#podemos ver que teníamos razón y si hay errores debido a que un apartamento entero debe de constar de dormitorio y un mínimo de 
#metros cuadrados para que sea funcional y tenga sentido

DataFrameDorm <- as.data.frame(table(barrio_sol$NumDormitorios))
ggplot(data=barrio_sol,aes(MetrosCuadrados))+geom_density(fill="#009999",color="#010101")+ggtitle("Diagrama de densidad de distribución de los tamaños de apartamentos en Sol")+theme(text=element_text(size=18))
ggplot(data=DataFrameDorm,aes(x=Var1, y=Freq))+geom_col(fill="#82E0AA",color="#010101")+ggtitle("Diagrama de densidad de distribución del número de dormitorios en Sol")+theme(text=element_text(size=18))

#Viendo la curva de densidad dada por los metros cuadrados y teniendo en cuenta que estamos hablando de apartamentos enteros podemos
#suponer que no tiene sentido un apartamento de menos de  metros cuadrados para que sea funcional, vamos a eliminar esos datos:

barrio_sol_act_metros <- subset(barrio_sol, MetrosCuadrados>=30)

#Viendo el histograma del número de dormitorios y volviendo a tener en cuenta que se trata de un apartamento entero, no tiene sentido
#que no tenga dormirotios por lo que vamos a eliminar los datos en los que el número de dormitorios sea nulo:

barrio_sol_final<- subset(barrio_sol_act_metros, NumDormitorios!=0)

#Comprobamos volviendo a dibujar la curva de densidad y el histograma:

DataFrameDormFinal <- as.data.frame(table(barrio_sol_final$NumDormitorios))
ggplot(data=barrio_sol_final,aes(MetrosCuadrados))+geom_density(fill="#009999",color="#010101")+ggtitle("Diagrama de densidad de distribución de los tamaños de apartamentos en Sol")+theme(text=element_text(size=18))
ggplot(data=DataFrameDormFinal,aes(x=Var1, y=Freq))+geom_col(fill="#82E0AA",color="#010101")+ggtitle("Diagrama de densidad de distribución del número de dormitorios en Sol")+theme(text=element_text(size=18))
```
